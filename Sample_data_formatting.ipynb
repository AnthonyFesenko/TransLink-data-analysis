```python
from google.transit import gtfs_realtime_pb2
import requests
import pandas as pd
import google.protobuf
from google.protobuf.json_format import MessageToJson
from google.protobuf.json_format import MessageToDict

#https://mayors-ic.github.io/examples/gtfs-example.html
#https://stackoverflow.com/questions/65816113/how-to-convert-gtfs-rt-trip-updates-data-to-a-dataframe
```

```python
feed = gtfs_realtime_pb2.FeedMessage()
api_key = {'apikey':['ceDOlkSthiAZ5pOZsfWL']}

response = requests.get('https://gtfsapi.translink.ca/v3/gtfsrealtime', params=api_key)
feed.ParseFromString(response.content)
trip_data = MessageToDict(feed)
trip_update = pd.json_normalize(trip_data['entity'])


response = requests.get('https://gtfsapi.translink.ca/v3/gtfsposition', params=api_key)
feed.ParseFromString(response.content)
position_data = MessageToDict(feed)
position_update = pd.json_normalize(position_data['entity'])


response = requests.get('https://gtfsapi.translink.ca/v3/gtfsalerts', params=api_key)
feed.ParseFromString(response.content)
service_data = MessageToDict(feed)
service_alerts = pd.json_normalize(service_data['entity'])
```
